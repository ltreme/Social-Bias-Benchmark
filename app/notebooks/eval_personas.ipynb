```python
import pandas as pd
import numpy as np
```

```python
personas = pd.read_csv('../../data/biasBench/personas.csv')
personas.head()
```

```python
total_personas = len(personas)
```

```python
total_personas = len(personas)

# Metriken für kategoriale Features
categorical_features = ['gender', 'ethnicity', 'religion', 'occupation', 'appearance', 'socioeconomic_status']
for col in categorical_features:
    print(f"\nVerteilung für {col}:")
    print(personas[col].value_counts(normalize=True) * 100) # In Prozent

# Metriken für numerische Features (age)
print("\nVerteilung für age:")
print(personas['age'].describe())

# Für Visualisierungen (optional, aber empfohlen)
import matplotlib.pyplot as plt
import seaborn as sns

# Balkendiagramme für kategoriale Features
for col in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=col, data=personas, order = personas[col].value_counts().index)
    plt.title(f'Verteilung von {col}')
    plt.xlabel('Anzahl')
    plt.ylabel(col)
    plt.show()

# Histogramm für age
plt.figure(figsize=(10, 6))
sns.histplot(personas['age'], kde=True, bins=10)
plt.title('Altersverteilung')
plt.xlabel('Alter')
plt.ylabel('Häufigkeit')
plt.show()

# Boxplot für age
plt.figure(figsize=(10, 6))
sns.boxplot(x=personas['age'])
plt.title('Altersverteilung (Boxplot)')
plt.xlabel('Alter')
plt.show()
```
